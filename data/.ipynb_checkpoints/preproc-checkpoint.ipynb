```python
#-*- coding:utf-8 -*-
import re #정규 표현식 모듈
import random
import math
```

```python
fp=open('7CM00042.txt', 'r', encoding='utf-16')
ff=open('chat.log','w',encoding='utf-8')
```

```python
Nstr = re.compile("\/[\W].*?")

cnt =0;
while True:
    line = fp.readline()

    if not line:
        break
        
    person = re.search('<u who="P\d">',line, re.I|re.S)
    if (person is not None):
        print (line[8:10]+ ": ", end='')
        
    personEnd = re.search('</u>',line, re.I|re.S)
    line = re.sub('<.*?>','',line ,0, re.I|re.S)
    line = re.sub(',','',line ,0, re.I|re.S)
    line = re.sub('\n','',line ,0, re.I|re.S)
    line = re.sub('::','',line ,0, re.I|re.S)

    
    if (personEnd is not None): #한사람 말 끝나면 줄넘김
        print(line)
        ff.write(line+"\n")
        cnt=cnt+1
    else: #말 이어서 하는중이면 줄넘김 없음
        print(line, end='')
        print(" ", end='')
        " ".join(line.split())
        ff.write(line)
        
    if cnt == 2:
        print("")
        cnt = 0

fp.close()
ff.close()
```
